## 1 Логнормальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html).

```python
from scipy.stats import lognorm
```

Випадкова велиина $Y$ має логнормальний розподіл, якщо $Y = \exp \left\{ X \right\}$, де $X$ - випадкова величина з нормального розподілу.

```python
mu = 0
sigma = 1
```

Параметри $s, loc, scale$ відповідають логнормальному розподілу зі щільністю $$f(x) = \frac{1}{s (x - loc) \sqrt{2 \pi}} \exp \left\{ -\frac{1}{2 s^2} \log^2 \left( \frac{x - loc}{scale} \right) \right\}.$$

```python
s = sigma
loc = 0
scale = np.exp(mu)
```

### 1.1 Щільність

```python
x1 = np.arange(-1, 5, 0.1)
```

Виведіть значення функції щільності логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm
mu = 0
sigma = 1
s = sigma
loc = 0
scale = np.exp(mu)
x1 = np.arange(-1, 5, 0.1)
a = lognorm.pdf(x1, s=s, loc=loc, scale=scale)
a

```

Зобразіть графічно значення функції щільності логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x2$.

```python
x2 = np.arange(-1, 5.01, 0.01)
```

```python
x2 = np.arange(-1, 5.01, 0.01)
plt.plot(x2, lognorm.pdf(x2, s=s, loc=loc, scale=scale), 'r', label='lognorm pdf x2')
plt.legend();

```

### 1.2 Функція розподілу

Виведіть значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

```python
a = lognorm.cdf(x1, s, loc=loc, scale=scale)
print(a)

```

Зобразіть графічно значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x2$.

```python
a2 = lognorm.cdf(x2, s=s, loc=loc, scale=scale)
plt.plot(x2, a2, 'b', label='lognorm cdf x2')
plt.legend();

```

## 2 Бета розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html).

```python
from scipy.stats import beta
```

```python
a = 2
b = 1.1
```

Параметри $loc, scale$ відповідають бета розподілу зі щільністю $$f(x) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta) \, scale} \left( \frac{x - loc}{scale} \right)^{\alpha - 1} \left( 1 - \frac{x - loc}{scale} \right)^{\beta - 1}.$$

```python
loc = 0
scale = 1
```

### 2.1 Щільність

```python
x1 = np.arange(-1, 2.1, 0.1)
```

Виведіть значення функції щільності бета розподілу з параметрами $a, b, loc, scale$ в точках вектора $x1$.

```python
from scipy.stats import beta
import numpy as np
a = 2
b = 1.1
loc = 0
scale = 1
x1 = np.arange(-1, 2.1, 0.1)
a = beta.pdf(x1,a,b,loc,scale)
print(a)

```

```python
plt.plot(x2, beta.pdf(x2, a, b, loc=loc, scale=scale), 'g', label="beta pdf x2")
plt.legend()
plt.show();

```

### 2.2 Функція розподілу

Зобразіть графічно функцію щільності, функцію розподілу та нормовану гістограму $100000$ випадкових величин з параметром `bins=50` разом з функцією щільності рівномірного розподілу з параметрами:
1. $a = 0, b = 2$
2. $a = 0, b = 5$
3. $a = 0, b = 10$
4. $a = -2, b = 2$
5. $a = -5, b = 5$

Зробіть шкали на відповідних графіках однаковими для кращого порівняння. Наприклад, усі графіки щільності повинні мати однакові границі як по осі $x$, так і по осі $y$.

```python
from scipy.stats import uniform

a = [0, 0, 0, -2, -5]
b = [2, 5, 10, 2, 5]
x = np.arange(-1, 2, 0.01)

for i in range(len(a)):
    plt.figure(figsize=(20, 5))
    fig.set_size_inches(15, 5)
    plt.subplots_adjust(hspace=0.4)
    
    loc = a[i]
    scale = b[i] - loc
    print("loc = {}, scale = {}".format(loc, scale))
    
    uni_pdf = uniform.pdf(x, loc=loc, scale=scale)
    uni_cdf = uniform.cdf(x, loc=loc, scale=scale)
    uni_rvs = uniform.rvs(loc=loc, scale=scale, size=100000)
    
    plt.subplot(1, 3, 1, ylim=(0, 1))
    plt.plot(x, uni_pdf, label='pdf')
    plt.legend()
    
    plt.subplot(1, 3, 2)
    plt.plot(x, uni_cdf, label='cdf')
    plt.legend()
    
    plt.subplot(1, 3, 3, xlim=(-1.2, 2.1), ylim=(0, 1))
    plt.plot(x, uni_pdf, 'r', label='pdf')
    plt.legend()
    plt.subplot(1, 3, 3, xlim=(-1.2, 2.1), ylim=(0, 1))
    plt.hist(uni_rvs, bins=50, label='rvs', normed=True)
    plt.legend()
    plt.show()

```

### 3.3 Нормальний розподіл

2. $\mu = 0, \sigma = 2$

4. $\mu = 0, \sigma = 1$
5. $\mu = -2, \sigma = 1$
6. $\mu = 2, \sigma = 1$

```python
from scipy.stats import norm

mu = [0, 0, 0, 0, -2, 2]
sig = [0.5, 2, 4, 1, 1, 1]
x = np.arange(-7, 7, 0.01)

for i in range(len(mu)):
    loc = mu[i]
    scale = sig[i]
    print("mu={}, sigma={}".format(loc, scale))
    
    n_pdf = norm.pdf(x, loc=loc, scale=scale)
    n_cdf = norm.cdf(x, loc=loc, scale=scale)
    n_rvs = norm.rvs(loc=loc, scale=scale, size=100000)
    
    fig = plt.figure(figsize=(20,5))
    fig.set_size_inches(15, 5)
    plt.subplots_adjust(hspace=0.4)

    plt.subplot(1, 3, 1, xlim=(-7.05, 7.05), ylim=(-0.05, 0.85))
    plt.plot(x, n_pdf, 'g', label='pdf')
    plt.legend()
    
    plt.subplot(1, 3, 2)
    plt.plot(x, n_cdf, 'g', label='cdf')
    plt.legend()
    
    plt.subplot(1, 3, 3, xlim=(-7.05, 7.05), ylim=(-0.05, 0.85))
    plt.plot(x, n_pdf, 'r', label='pdf')
    plt.legend()
    plt.subplot(1, 3, 3, xlim=(-7.05, 7.05), ylim=(-0.05, 0.85))
    plt.hist(n_rvs, bins=50, label='rvs', normed=True)
    plt.legend()
    plt.show()

```

**Зауваження.** Математичне сподівання та дисперсія відповідного нормального розподілу збігається з математичним сподіванням та дисперсією біноміального розподілу.

```python
from scipy.stats import norm
from scipy.stats import binom
n=100
list_p=[0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]
for p in list_p:
    mu = n*p #мат.сподівання
    sigma = (n*p*(1-p))**0.5 #D=sigma**2
    loc = mu
    scale = sigma
    x2 = np.arange(0, 100, 0.1)
    x2g = np.arange(0, 100, 1)
    y2 = norm.pdf(x2, loc=loc, scale=scale)
    y2_ =norm.cdf(x2, loc=loc, scale=scale)
    fig, axs = plt.subplots(1,2, figsize=(20,5))
    yg = binom.pmf(x2g, n, p)
    yg_ = binom.cdf(x2g, n, p)
    
    axs[0].plot(x2, y2, 'r--', x2g, yg, 'b')
    axs[0].set_title('pdf')
    axs[1].plot(x2, y2_, 'r--', x2g, yg_, 'b')
    axs[1].set_title('cdf')

```
